<h3>Tokenizing</h3>

```python
tokenizing = pd.read_csv('Tokenizing.csv', encoding='latin-1')
tokenizing.head()
```

<h3>Phrase Detection</h3>

```python
phrase_detection = tokenizing

phrase_detection['Bahan_1'] = tokenizing['Bahan_1'].str.replace(' ', '-')
phrase_detection['Bahan_2'] = tokenizing['Bahan_2'].str.replace(' ', '-')
phrase_detection['Bahan_3'] = tokenizing['Bahan_3'].str.replace(' ', '-')
phrase_detection['Bahan_4'] = tokenizing['Bahan_4'].str.replace(' ', '-')
phrase_detection['Bahan_5'] = tokenizing['Bahan_5'].str.replace(' ', '-')
phrase_detection.head()
```

<h3>TF-IDF</h3>

```python
phrase_detection['Token'] = phrase_detection.apply(lambda x: x['Bahan_1'] + ' ' + x['Bahan_2'] + ' ' + x['Bahan_3'] + ' ' + x['Bahan_4'] + ' ' + x['Bahan_5'], axis=1)
phrase_detection.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


tf_idf = phrase_detection
vectorizer = TfidfVectorizer(token_pattern=r"(?u)\b[\w&.\-]+\b")
X = vectorizer.fit_transform(tf_idf['Token'])
y = tf_idf['Nama_Resep']
# y_target = y['Nama_Resep']
assert X.shape[0] == y.shape[0]
terms = vectorizer.get_feature_names_out()

for idx, recipe in enumerate(y):
    # Mendapatkan nilai tfidf
    tfidf_values = X[idx].toarray()[0]

    # Mengurutkan indeks Resep sesuai abjad
    sorted_indices = tfidf_values.argsort()[::-1]

    print(f"Resep: {recipe}")
    print("Istilah yang paling berkorelasi:")
    for index in sorted_indices[:5]:
        term = terms[index]
        tfidf = tfidf_values[index]
        print(f"- Term: {term}, TF-IDF: {tfidf}")
    print("----------")
```
